```python
# Converting the obj of release_date to date_time
df_copy["Release_Date"] = pd.to_datetime(df_copy["Release_Date"])
print(df_copy["Release_Date"].dtypes)
```

```python
# Changing the datatype of datetime
df_copy["Release_Date"] = df_copy["Release_Date"].dt.year
df_copy["Release_Date"].dtypes
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
# dropping overview, original_language, poster_url
cols = ["Overview", "Original_Language", "Poster_Url"]

df_copy.drop(columns= cols, inplace=True)
df_copy.columns
```

```python
df_copy.head()
```

```python
# categorizing vote_average column
def categorize_column(df_copy, col, label):
    edges = [df_copy[col].describe()["min"],
             df_copy[col].describe()["25%"],
             df_copy[col].describe()["50%"],
             df_copy[col].describe()["75%"],
             df_copy[col].describe()["max"]]
    df_copy[col] = pd.cut(df_copy[col], edges, labels= label, duplicates='drop')
    return df_copy

labels = ["not_popular", "below_avg", "average", "popular"]
categorize_column(df_copy, "Vote_Average", label= labels)

# Confirming changes
df_copy["Vote_Average"].unique()
```

```python
df_copy.head()
```

```python
# exploring columns

df_copy["Vote_Average"].value_counts()
```

```python
df_copy.isnull().sum()
```

```python
#Dropping null values
df_copy.dropna(inplace=True)
df_copy.isna().sum()
```

```python
df_copy.head()
```

```python
# splitting the genre 
df_copy["Genre"] = df_copy["Genre"].str.split(", ")

# explode the lists
df_copy = df_copy.explode("Genre").reset_index(drop=True)
df_copy.head()
```

```python
# casting column into category
df_copy["Genre"]= df_copy["Genre"].astype("category")

# confirming changes
df_copy["Genre"].dtypes
```
